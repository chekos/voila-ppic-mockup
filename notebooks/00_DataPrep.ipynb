```python
data = pd.read_csv(INTERIM_DATA / 'nhgis0005_csv/nhgis0005_csv' / 'nhgis0005_ds232_2017_county.csv', encoding = 'latin1')
```

```python
data.head()
```

```python
data.head().T
```

```python
data = data[['GISJOIN', 'AG90E015', 'AG90E032']]
```

```python
data.columns = ['GISJOIN', 'male', 'female']
```

```python
data = data.melt(id_vars='GISJOIN', var_name='sex', value_name='count')
```

```python
ca = ca[['GISJOIN', 'geometry']]
```

```python
data.set_index('GISJOIN', inplace = True)
ca.set_index('GISJOIN', inplace = True)
```

```python
geo = ca.join(data)
```

```python
geo.reset_index().to_file(PROCESSED_DATA / 'CA_data.shp', index = False)
```
